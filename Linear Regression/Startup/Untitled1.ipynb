```python
import numpy as np
import pandas as pd
```

```python
md = pd.read_csv("50_Startups.csv")
md.head()
```

```python
md.State.unique()
```

```python
dummy = pd.get_dummies(md.State, drop_first = True)
dummy.head()
```

```python
md = pd.concat([md, dummy], axis =1)
md.head()
```

```python
md.drop('State', axis =1, inplace = True)
md.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
md_train, md_test = train_test_split(md, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
md_train.corr()
```

```python
md_train["Profit"].hist()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
num_list = ["R&D Spend", "Administration", "Marketing Spend", "Profit"]
md_train[num_list] = scaler.fit_transform(md_train[num_list])
```

```python
md_train.head()
```

```python
y_train = md_train.pop("Profit")
X_train = md_train
```

```python
y_train.head()
```

```python
X_train.head()
```

```python
import statsmodels.api as sm
```

```python
X_train_lm1 = sm.add_constant(X_train)
X_train_lm1.head()
```

```python
lr_1 = sm.OLS(y_train, X_train_lm1).fit()
lr_1.params
```
